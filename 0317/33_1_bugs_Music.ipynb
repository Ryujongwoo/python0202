```python
# 벅스 뮤직 Top 100 크롤링
import requests
from bs4 import BeautifulSoup
targetSite = 'https://music.bugs.co.kr/chart'
```

```python
request = requests.get(targetSite)
# print(request)
html = request.text
# print(html)
soup = BeautifulSoup(html, 'html.parser')
# print(soup)
```

```python
# 노래 제목 크롤링
titles = soup.findAll('p', {'class': 'title'})
# print(titles)
print(len(titles))
for title in titles:
    print(title.text.strip())
```

```python
# 가수 이름 크롤링
artists = soup.findAll('p', {'class': 'artist'})
# print(artists)
print(len(artists))
for artist in artists:
    print(artist.text.strip().split('\n')[0])
```

```python
bugs = [] # 노래 제목과 가수를 하나의 리스트로 만들어 저장할 빈 리스트를 선언한다.
titles = soup.findAll('p', {'class': 'title'})
artists = soup.findAll('p', {'class': 'artist'})
for i in range(len(titles)):
    bugs.append([titles[i].text.strip(), artists[i].text.strip().split('\n')[0]])
# print(bugs)

from datetime import datetime as dt
print(dt.now(), '현재')
for i in range(len(bugs)):
    print('{0:3d}위. {1} - {2}'.format(i + 1, bugs[i][0], bugs[i][1]))
```

```python
from datetime import datetime as dt

# file = open('bugsTop100.txt', 'w') 처럼 파일을 열어서 저장할 때 'cp949' codec can't encode character와 같은 에러가
# 발생되면 아래와 같이 열어서 UTF-8(유니코드)로 인코딩 시켜서 저장해야 한다.
file = open('bugsTop100.txt', 'w', -1, 'UTF-8')
file.write(str(dt.now()) + '\n')
for i in range(len(bugs)):
    data = '{0:3d}위. {1} - {2}'.format(i + 1, bugs[i][0], bugs[i][1])
    # print(data)
    file.write(data + '\n')
file.close()
print('bugsTop100.txt 파일에 쓰기 완료')
```

```python
file = open('bugsTop100.txt', 'r', -1, 'UTF-8')
lines = file.readlines()
print('벅스뮤직 TOP 100')
for line in lines:
    print(line, end = '')
print()
file.close()
```
