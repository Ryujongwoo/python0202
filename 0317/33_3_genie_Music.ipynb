```python
# 지니 뮤직 Top 100 크롤링
import requests
from bs4 import BeautifulSoup
#'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20200317&hh=20&rtm=Y&pg=1' =>   1~ 50위
#'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20200317&hh=20&rtm=Y&pg=2' =>  51~100위
#'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20200317&hh=20&rtm=Y&pg=3' => 101~150위
#'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20200317&hh=20&rtm=Y&pg=4' => 151~200위
header = {'User-Agent': 'Mozilla/5.0 (compatible, MSIE 11, Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
url = 'https://www.genie.co.kr/chart/top200?ditc=D&ymd=20200317&hh=20&rtm=Y&pg=' # 공통 부분
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    print(targetSite)
```

```python
# 노래 제목 크롤링
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers = header)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    titles = soup.findAll('a', {'class': 'title ellipsis'})
    print(len(titles))
    for title in titles:
        print(title.text.strip())
    print('*' * 75) 
```

```python
# 가수 이름 크롤링
for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers = header)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    artists = soup.findAll('a', {'class': 'artist ellipsis'})
    print(len(artists[5:]))
    # 그래프에 나타나는 1~5위 가수 이름이 같이 크롤링되므로 슬라이싱을 이용해서 그래프의 가수를 제외시킨다.
    for artist in artists[5:]:
        print(artist.text.strip().split('\n')[0])
    print('*' * 75)
```

```python
genie = [] # 노래 제목과 가수를 하나의 리스트로 만들어 저장할 빈 리스트를 선언한다.

from datetime import datetime as dt
print(dt.now(), '현재')

for i in range(1, 5):
    targetSite = '{}{}'.format(url, i)
    request = requests.get(targetSite, headers = header)
    html = request.text
    soup = BeautifulSoup(html, 'html.parser')
    titles = soup.findAll('a', {'class': 'title ellipsis'})   # 50개씩
    artists = soup.findAll('a', {'class': 'artist ellipsis'}) # 55개씩
    for i in range(len(titles)):
        genie.append([titles[i].text.strip(), artists[i + 5].text.strip().split('\n')[0]])

for i in range(len(genie)):
    print('{0:3d}위. {1} - {2}'.format(i + 1, genie[i][0], genie[i][1]))
```

```python
from datetime import datetime as dt

file = open('genieTop200.txt', 'w', -1, 'UTF-8')
file.write(str(dt.now()) + '\n')
for i in range(len(genie)):
    data = '{0:3d}위. {1} - {2}'.format(i + 1, genie[i][0], genie[i][1])
    # print(data)
    file.write(data + '\n')
file.close()
print('genieTop200.txt 파일에 쓰기 완료')
```
