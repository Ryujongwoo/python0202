```python
# 3이상인 홀수를 입력받는다.
while True:
    n = int(input('3이상인 홀수를 입력하세요 : '))
    if n >= 3 and n % 2 == 1:
        break
    print('3이상인 홀수를 입력해야 합니다.')
# =============================================

# n행 n열 리스트를 선언한다.
a = [[0] * n for i in range(n)]
# 최초의 0이 채워질 위치를 계산한다.
i = 0
j = n // 2
# print('i = {}, j = {}'.format(i, j))

for k in range(1, n ** 2 + 1):
    a[i][j] = k
    if k % n == 0:
        i += 1
    else:
        i -= 1
        if i < 0:
            i = n - 1
        j += 1
        if j > n - 1:
            j = 0
    
for i in range(n):
    for j in range(n):
        print('{0:3d} '.format(a[i][j]), end = '')
    print()
```
