함수의 구조  
def 함수이름([인수, ...]): # 인수는 함수가 처리할 데이터를 의미하며 필요없으면 생략 가능하다.  
&nbsp;&nbsp;&nbsp;&nbsp;함수가 실행할 문장  
&nbsp;&nbsp;&nbsp;&nbsp;...  
&nbsp;&nbsp;&nbsp;&nbsp;[return 함수의 실행 결과] # 함수의 실행 결과가 없으면 return을 생략 가능하다.

```python
# 가장 일반적인 함수의 모양
# a, b 2개의 정수 값을 인수로 넘겨받아 덧셈을 실행하고 실행된 결과를 리턴하는 함수
def add(a, b):
    print('add() 함수가 실행됩니다.')
    print('a = {}, b = {}'.format(a, b))
    c = a + b
    print('{} + {} = {}'.format(a, b, c))
    # return을 만나거나 함수가 종료되면 함수를 호출한 곳으로 돌아간다.
    # return을 만나면 리턴값을 가지고 돌아간다.
    return c
```

```python
# 함수를 실행하려면 실행하려는 함수의 이름을 적어주면 된다.
print('함수가 실행된 결과 : {}'.format(add(5, 3)))
```

```python
# 입력값이 없는 함수, 처리할 데이터(인수)를 넘겨받지 않는 함수.
def func1():
    print('호출된 함수에서 출력 => func1() 함수가 실행됩니다.')
    return 'func1() 함수가 실행됩니다.'
```

```python
print('함수를 호출한 곳에서 출력 => {}'.format(func1()))
```

결과값이 없는 함수, return이 생략된 함수

```python
def func2(a, b):
    print('호출된 함수에서 출력 => func2() 함수가 실행됩니다.')
    print('{} + {} = {}'.format(a, b, a + b))
```

```python
func2(4, 7)
```

오버로딩 => 1개의 파일에서 같은 이름을 가지는 함수를 여러개 만들어 사용할 수 있다.  
오버로딩된 함수는 인수의 개수로 함수를 구별하고 인수의 개수도 같다면 인수의 데이터 타입으로 구별한다.  
파이썬은 오버로딩이 지원되지 않는다. => 오버로딩 효과를 내기 위해 가변 인자를 가지는 함수를 사용한다.

```python
# 파이썬은 같은 이름을 가지는 함수를 만들면 맨 마지막에 정의한 함수만 유효하고 앞에 정의한 함수는
# 뒤에 정의한 함수로 덮어쓰기 된다.
def func2(a, b, c):
    print('호출된 함수에서 출력 => func2() 함수가 실행됩니다.')
    print('{} + {} + {} = {}'.format(a, b, c, a + b + c))
```

```python
# func2(4, 7) # 에러 발생
func2(4, 7, 10)
```

```python
# 인수의 개수가 가변적인 함수 => 가변 인수를 사용하는 함수 => 가변 인수는 *변수명 형태로 사용한다.
# 변수명 앞에 '*'을 붙여주면 가변 인수로 취급되고 넘어오는 데이터를 튜플로 받는다.
def func3(*args):
    print('func3() 함수가 실행됩니다.')
    print(type(args))
    print(args)
    print('func3() 함수로 넘어온 인수의 개수 : {}'.format(len(args)))
    
    total = 0
    for arg in args:
        total += arg
    return total
```

```python
result = func3(1, 2, 3, 100, 5, 6, 7, 8, 9)
print('result = {}'.format(result))
```

```python
# 디폴트 인수를 사용하는 함수 => 변수명 = 디폴트값(기본값)
# 함수가 호출될 때 인수가 넘어노면 넘어온 값으로 함수를 실행하고 인수가 넘어오지 않으면 지정된
# 디폴트값으로 함수를 실행한다.
# 디폴트 인수는 부분적으로 사용할 수 없다. => 모든 인수에 사용하거나 모든 인수에 사용하지 않아야
# 한다.
def func4(a = 5, b = 8):
    print('func4() 함수가 실행됩니다.')
    result = a - b
    print('{} - {} = {}'.format(a, b, a - b))
    return result
```

```python
print(func4(101, 10))
print(func4(101))
print(func4())
# 함수를 호출할 때 인수로 전달되는 값을 직접 지정할 수 있다.
print(func4(999, 1))
print(func4(b = 999, a = 1))
```

```python
# 넘겨받는 인수를 딕셔너리 형태로 반환하는 함수 => 딕셔너리 형태로 반환하는 함수는 **변수명 형태로
# 사용한다.
# 넘겨받은 인수를 딕셔너리 형태로 변환해서 '**'을 붙여준 변수에 저장한다.
def func5(**args):
    print('func5() 함수가 실행됩니다.')
    print(type(args))
    print(args)
    return args
```

```python
# '**'이 앞에 붙은 인수는 key = value 형태로 인수를 전달해야 한다.
func5(apple = 1000)
func5(apple = 1000, banana = 2000)
print(func5(apple = 1000, banana = 2000, melon = 10000))
```

```python
# return을 튜플로 하면 return 값이 2개 이상인 것 같은 효과를 낸다.
def func6(a, b):
    temp = a
    a = b
    b = temp
    return a, b # retuen을 튜플로 한다.
```

```python
a = 3
b = 4
print('a = {}, b = {}'.format(a, b))
a, b = b, a
print('a = {}, b = {}'.format(a, b))
a, b = func6(a, b)
print('a = {}, b = {}'.format(a, b))
```
